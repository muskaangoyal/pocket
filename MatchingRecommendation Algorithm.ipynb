```python
data = pd.read_csv('data.csv')
data['userID'] = data.index
data.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
data.head()
```

```python
Ratings = data.iloc[:, np.arange(7, 29)]
Ratings.drop('Transportation', axis=1, inplace=True)
Ratings.head()
```

```python
Ratings_melted = pd.melt(Ratings, id_vars='userID', value_vars=Ratings.columns.values[:20])
Ratings_melted.head()
```

```python
# Replacing NaN Average
Ratings = Ratings.fillna(Ratings.mean(axis=0))
Ratings.head()
```

```python
cosine = cosine_similarity(Ratings.drop('userID', axis=1))
np.fill_diagonal(cosine, 0)
similarity_with_tip = pd.DataFrame(cosine,index=Ratings.index)
similarity_with_tip.columns=Ratings.index
similarity_with_tip
```

```python
def find_n_neighbours(df,n):
    order = np.argsort(df.values, axis=1)[:, :n]
    df = df.apply(lambda x: pd.Series(x.sort_values(ascending=False)
           .iloc[:n].index, 
          index=['top{}'.format(i) for i in range(1, n+1)]), axis=1)
    return df
```

```python
# top 5 neighbours for each user
sim_user_5_m = find_n_neighbours(similarity_with_tip,5)
sim_user_5_m
```

```python
def get_user_similar_tips(user1, user2):
    common_tips = Ratings_melted[Ratings_melted['userID'] == user1].merge(
        Ratings_melted[Ratings_melted.userID == user2],
        on = "variable",
        how = "inner" )
    return common_tips# .merge( tips, on = 'tipID' )
```

```python
opinion = get_user_similar_tips(37,86)
opinion
```

```python
import operator

def recommended(user1): 
    sim = sim_user_5_m.loc[user1, :].values.tolist()
    dic = {}
    for u in sim: 
        u_list = Ratings_melted[Ratings_melted['userID'] == u]
        u_list = u_list[u_list['value'] == 1.0]['variable'].unique()
        for t in u_list:
            if t in dic:
                dic[t] += 1
            else:
                dic[t] = 1
    sorted_x = sorted(dic.items(), key=operator.itemgetter(1), reverse=True)
    return [it[0] for it in sorted_x[:3]]
```

```python
recommended(1)
```

```python
user = int(input("Enter the user id for whom you want to recommend some tips: "))
for i in recommended(user):
    print(i)
```
